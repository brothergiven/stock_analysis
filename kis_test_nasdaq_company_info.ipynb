### NASDAQ 회사 정보 저장

```python
import pandas as pd
import os
```

```python
colnames = [
    "ncod", "exid", "excd", "exnm", "symb", "rsym", "knam", "enam",
    "stis", "curr", "zdiv", "ztyp", "base", "bnit", "anit",
    "mstm", "metm", "isdr", "drcd", "icod", "sjong", "ttyp", "etyp", "ttyp_sb"
]
```

```python
df = pd.read_csv("./files/NASMST.csv", header=None, encoding="cp949", names = colnames)
df
```

```python
df = df[[
    "symb", "knam", "enam", "exnm", "stis",
    "curr", "base", "isdr", "sjong"
]]
```

```python
df = df.rename(columns={
    "symb": "종목코드",
    "knam": "종목명",
    "enam": "영문명",
    "exnm": "거래소",
    "stis": "상품유형",
    "curr": "통화",
    "base": "기준가",
    "isdr": "예탁증서여부",
    "sjong": "지수구성"
})

df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df
```
